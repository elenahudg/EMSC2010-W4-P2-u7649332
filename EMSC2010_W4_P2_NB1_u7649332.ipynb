This record is referred to as the *Keeling curve* in honour of Charles Keeling who started the monitoring program and ran it until his death in 2005.

```python
## Importing all important packages:
# matplotlib for plotting
# numpy for working with numbers
# pandas for reading in data

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Read in the csv data using pandas. Data found from  Dr. Xin Lan, NOAA/GML (gml.noaa.gov/ccgg/trends/) and Dr. Ralph Keeling, Scripps Institution of Oceanography (scrippsco2.ucsd.edu/). The CSV file was edited to remove information in the headers, and only contain the relevant data.

```python
df=pd.read_csv("co2_mm_mlo.csv")
```

```python
df.head() # check the headers to know how to call each column
```

```python
time=df["decimal date"] # create a variable called time with the decimal date data
co2=df["average"] # create a variable called co2 with the average CO2 data
```

```python
plt.plot(time,co2) # using the created variables to make the plot
## Use the rcParams function to allow some LaTeX capabilities (for the subscript 2 in CO2)
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)
plt.xlabel("Time (months)") # add an x-axis label
plt.ylabel("CO$_2$ (ppm)") # add a y-axis label
plt.title('Mauna Loa Monthly Mean CO$_2$ Data') # add a title
```
